# Enterprise Change Intelligence Platform
## AI-Powered Impact Analysis for ERP/Product Management

---

## 1. Executive Summary

This platform provides **AI-driven change impact analysis** for enterprise systems. It maintains a persistent knowledge base of enterprise artifacts (requirements, design docs, API specs, release notes) and uses semantic search + LLM analysis to:

* Analyze new business requirements
* Identify impacted modules, screens, APIs, reports, workflows
* Generate structured impact assessments with traceability
* Build enterprise memory for release intelligence

**Core Principles:**
* **Modularity**: Clean separation of concerns (ingestion, retrieval, analysis, UI)
* **Scalability**: Vector DB + chunking strategy for millions of documents
* **Extensibility**: Pluggable AI providers (Gemini/OpenAI/Claude/Databricks)
* **Traceability**: Source attribution for every insight
* **Enterprise-grade**: Metadata tracking, versioning, audit trail

## 5. Core Modules & Services

### 5.1 Ingestion Service

**Purpose:** Transform raw documents into searchable, semantically-indexed chunks.

| Component | Responsibility |
|-----------|---------------|
| **Document Loader** | Parse PDF/DOCX/PPTX/XLSX/TXT/JSON into text |
| **Chunking Engine** | Split text into semantic chunks (500-1000 tokens) with overlap |
| **Embedding Service** | Generate 1024-dim vectors via Databricks BGE/GTE |
| **Metadata Extractor** | Extract: doc_name, module, domain, release, upload_date, source_path, doc_type |
| **Repository Scanner** | Recursive folder traversal, change detection |

**Key Design Decisions:**
* **Chunking Strategy:** Hybrid approach - respect section boundaries (markdown headers, page breaks) while enforcing max token limits
* **Metadata Schema:** Store metadata both in Delta table AND as filters in Vector Search for fast filtering
* **Idempotency:** Track document hashes to avoid re-processing unchanged files

---

### 5.2 Storage Layer

**Purpose:** Durable, scalable storage for vectors, metadata, and raw documents.

| Store | Technology | Schema |
|-------|------------|--------|
| **Vector Store** | Databricks Vector Search (Storage-Optimized) | `doc_id, chunk_id, text, embedding, metadata` |
| **Metadata Store** | Delta Table (Unity Catalog) | `doc_id, doc_name, module, domain, release, upload_date, source_path, doc_type, chunk_count, status` |
| **Repository** | Local filesystem or S3 | Original documents in structured folders |

**Storage-Optimized vs Standard Endpoint:**
* Use **Storage-Optimized** for cost-effective storage (1B+ vectors, 7x cheaper, 300-500ms latency)
* Use **Standard** only if <100ms latency is critical

**Delta Table Schema for Metadata:**
```sql
CREATE TABLE IF NOT EXISTS main.change_intel.document_metadata (
  doc_id STRING PRIMARY KEY,
  doc_name STRING,
  module STRING,
  domain STRING,
  release STRING,
  upload_date TIMESTAMP,
  source_path STRING,
  doc_type STRING,
  chunk_count INT,
  status STRING,
  file_hash STRING,
  indexed_at TIMESTAMP
)
TBLPROPERTIES (
  delta.enableChangeDataFeed = true
);
```

---

### 5.3 Retrieval Service

**Purpose:** Find relevant context from enterprise knowledge base.

| Component | Approach |
|-----------|----------|
| **Semantic Retriever** | Pure vector similarity (ANN search) |
| **Hybrid Retriever** | Vector + BM25 keyword matching (best for technical terms, IDs, acronyms) |
| **Reranker** | Post-retrieval reranking using cross-encoder or LLM scoring |
| **Query Optimizer** | Query expansion, decomposition, domain-specific filtering |

**Retrieval Strategy:**
```
User Query → Query Optimizer (expand, add domain context)
           ↓
       Hybrid Search (top 50 chunks)
           ↓
       Reranker (top 10 chunks)
           ↓
       LLM Analysis
```

**When to Use Hybrid vs Semantic:**
* **Hybrid:** Queries with exact terms (error codes, API names, screen IDs, SKUs)
* **Semantic:** Natural language questions, conceptual queries

## 8. Data Flow

### Indexing Flow
```
1. User uploads document OR Repository scanner detects new file
   ↓
2. Document Loader extracts text
   ↓
3. Metadata Extractor parses: module, domain, release, doc_type
   ↓
4. Chunking Engine splits into 500-1000 token chunks (with overlap)
   ↓
5. Embedding Service generates 1024-dim vectors (Databricks BGE/GTE)
   ↓
6. Write to Delta table: main.change_intel.document_chunks
   |
   ├── Columns: doc_id, chunk_id, text, module, domain, release, doc_type, source_path
   └── Vector Search auto-syncs and computes embeddings
   ↓
7. Update metadata table: main.change_intel.document_metadata
   └── Status: indexed, chunk_count, indexed_at
```

### Query Flow
```
1. User enters requirement text + optionally filters (domain, module, release)
   ↓
2. Query Optimizer: expand query, add domain context
   ↓
3. Hybrid Retrieval:
   ├── Vector Search: query_index() with filters
   └── Returns top 50 chunks with similarity scores
   ↓
4. Reranker (optional): Cross-encoder rescoring → top 10 chunks
   ↓
5. LLM Orchestrator:
   ├── Build prompt: system prompt + requirement + retrieved context
   ├── Call LLM (Databricks/Gemini/OpenAI/Claude)
   └── Parse structured output (JSON schema)
   ↓
6. Output Formatter: Validate and enrich output
   └── Add source document links, confidence scores
   ↓
7. Display in UI + Store in query history
```

## 9. Key Design Principles

1. **Modularity:** Each service is independently testable and replaceable
2. **Extensibility:** Add new document types, AI providers, or domains without core changes
3. **Traceability:** Every insight links back to source documents/chunks
4. **Idempotency:** Re-indexing same document is safe (hash-based deduplication)
5. **Scalability:** Storage-Optimized Vector Search handles 1B+ vectors
6. **Maintainability:** Clear separation of concerns, comprehensive logging
7. **Enterprise-grade:** Audit trails, RBAC, SSO integration in future phases

````python
# src/ingestion/document_loader.py
"""Multi-format document loader with metadata extraction."""

from abc import ABC, abstractmethod
from pathlib import Path
from typing import Dict, List, Optional
import hashlib
import mimetypes

# Document parsing libraries
try:
    from pypdf import PdfReader
    from docx import Document
    from pptx import Presentation
    from openpyxl import load_workbook
except ImportError:
    print("Install: pip install pypdf python-docx python-pptx openpyxl")

class DocumentLoader(ABC):
    """Base class for document loaders."""
    
    @abstractmethod
    def load(self, file_path: Path) -> Dict[str, any]:
        """Load document and return text + metadata."""
        pass
    
    def compute_hash(self, file_path: Path) -> str:
        """Compute SHA256 hash for deduplication."""
        sha256 = hashlib.sha256()
        with open(file_path, "rb") as f:
            for chunk in iter(lambda: f.read(8192), b""):
                sha256.update(chunk)
        return sha256.hexdigest()

class PDFLoader(DocumentLoader):
    """Load PDF documents."""
    
    def load(self, file_path: Path) -> Dict[str, any]:
        reader = PdfReader(str(file_path))
        text = ""
        for page in reader.pages:
            text += page.extract_text() + "\n\n"
        
        return {
            "text": text.strip(),
            "page_count": len(reader.pages),
            "metadata": reader.metadata or {},
            "file_hash": self.compute_hash(file_path)
        }

class DOCXLoader(DocumentLoader):
    """Load DOCX documents."""
    
    def load(self, file_path: Path) -> Dict[str, any]:
        doc = Document(str(file_path))
        text = "\n\n".join([para.text for para in doc.paragraphs if para.text.strip()])
        
        return {
            "text": text.strip(),
            "paragraph_count": len(doc.paragraphs),
            "metadata": doc.core_properties.__dict__,
            "file_hash": self.compute_hash(file_path)
        }

class PPTXLoader(DocumentLoader):
    """Load PPTX presentations."""
    
    def load(self, file_path: Path) -> Dict[str, any]:
        prs = Presentation(str(file_path))
        text = ""
        
        for slide in prs.slides:
            for shape in slide.shapes:
                if hasattr(shape, "text"):
                    text += shape.text + "\n"
            text += "\n"
        
        return {
            "text": text.strip(),
            "slide_count": len(prs.slides),
            "metadata": {},
            "file_hash": self.compute_hash(file_path)
        }

class XLSXLoader(DocumentLoader):
    """Load XLSX spreadsheets."""
    
    def load(self, file_path: Path) -> Dict[str, any]:
        wb = load_workbook(str(file_path), read_only=True)
        text = ""
        
        for sheet_name in wb.sheetnames:
            sheet = wb[sheet_name]
            text += f"\n\n=== Sheet: {sheet_name} ===\n\n"
            
            for row in sheet.iter_rows(values_only=True):
                row_text = "\t".join([str(cell) if cell is not None else "" for cell in row])
                if row_text.strip():
                    text += row_text + "\n"
        
        return {
            "text": text.strip(),
            "sheet_count": len(wb.sheetnames),
            "metadata": {},
            "file_hash": self.compute_hash(file_path)
        }

class TXTLoader(DocumentLoader):
    """Load plain text files."""
    
    def load(self, file_path: Path) -> Dict[str, any]:
        with open(file_path, "r", encoding="utf-8", errors="ignore") as f:
            text = f.read()
        
        return {
            "text": text.strip(),
            "line_count": text.count("\n"),
            "metadata": {},
            "file_hash": self.compute_hash(file_path)
        }

class DocumentLoaderFactory:
    """Factory for creating appropriate document loader."""
    
    _loaders = {
        ".pdf": PDFLoader,
        ".docx": DOCXLoader,
        ".pptx": PPTXLoader,
        ".xlsx": XLSXLoader,
        ".txt": TXTLoader,
    }
    
    @classmethod
    def get_loader(cls, file_path: Path) -> DocumentLoader:
        """Get appropriate loader based on file extension."""
        ext = file_path.suffix.lower()
        loader_class = cls._loaders.get(ext)
        
        if loader_class is None:
            raise ValueError(f"Unsupported file type: {ext}")
        
        return loader_class()
    
    @classmethod
    def load_document(cls, file_path: Path) -> Dict[str, any]:
        """Load document using appropriate loader."""
        loader = cls.get_loader(file_path)
        result = loader.load(file_path)
        
        # Add common metadata
        result["file_name"] = file_path.name
        result["file_path"] = str(file_path)
        result["file_size"] = file_path.stat().st_size
        result["file_type"] = file_path.suffix.lower()[1:]  # Remove dot
        
        return result

# Example usage
if __name__ == "__main__":
    # Demo
    print("Document Loader Factory initialized!")
    print(f"Supported formats: {list(DocumentLoaderFactory._loaders.keys())}")
````

````python
# src/storage/vector_store.py
"""Vector Search interface for document retrieval."""

from typing import List, Dict, Optional
from databricks.sdk import WorkspaceClient
from databricks.vector_search.client import VectorSearchClient
import json

class VectorStore:
    """Wrapper for Databricks Vector Search operations."""
    
    def __init__(
        self,
        endpoint_name: str,
        index_name: str,
        catalog: str,
        schema: str,
        chunks_table: str
    ):
        self.endpoint_name = endpoint_name
        self.index_name = index_name
        self.catalog = catalog
        self.schema = schema
        self.chunks_table = chunks_table
        
        self.workspace_client = WorkspaceClient()
        self.vs_client = VectorSearchClient()
        
    def create_endpoint(self, endpoint_type: str = "STORAGE_OPTIMIZED"):
        """Create Vector Search endpoint."""
        try:
            endpoint = self.workspace_client.vector_search_endpoints.create_endpoint(
                name=self.endpoint_name,
                endpoint_type=endpoint_type
            )
            print(f"Endpoint '{self.endpoint_name}' created successfully!")
            return endpoint
        except Exception as e:
            if "already exists" in str(e).lower():
                print(f"Endpoint '{self.endpoint_name}' already exists.")
            else:
                raise e
    
    def create_delta_sync_index(
        self,
        primary_key: str = "chunk_id",
        text_column: str = "text",
        embedding_model: str = "databricks-gte-large-en",
        pipeline_type: str = "TRIGGERED"
    ):
        """Create Delta Sync index with managed embeddings."""
        
        source_table = f"{self.catalog}.{self.schema}.{self.chunks_table}"
        
        try:
            index = self.workspace_client.vector_search_indexes.create_index(
                name=self.index_name,
                endpoint_name=self.endpoint_name,
                primary_key=primary_key,
                index_type="DELTA_SYNC",
                delta_sync_index_spec={
                    "source_table": source_table,
                    "embedding_source_columns": [
                        {
                            "name": text_column,
                            "embedding_model_endpoint_name": embedding_model
                        }
                    ],
                    "pipeline_type": pipeline_type
                }
            )
            print(f"Index '{self.index_name}' created successfully!")
            return index
        except Exception as e:
            if "already exists" in str(e).lower():
                print(f"Index '{self.index_name}' already exists.")
            else:
                raise e
    
    def sync_index(self):
        """Manually trigger index sync for TRIGGERED pipeline."""
        self.workspace_client.vector_search_indexes.sync_index(
            index_name=self.index_name
        )
        print(f"Index sync triggered for '{self.index_name}'.")
    
    def query(
        self,
        query_text: str,
        num_results: int = 50,
        filters: Optional[Dict[str, any]] = None,
        columns: Optional[List[str]] = None,
        query_type: str = "HYBRID"  # or "ANN"
    ) -> List[Dict[str, any]]:
        """Query the vector index.
        
        Args:
            query_text: Natural language query
            num_results: Number of results to return
            filters: Filter conditions (e.g., {"domain": "ERP", "module": "Order"})
            columns: Columns to return
            query_type: "HYBRID" (semantic + keyword) or "ANN" (semantic only)
        
        Returns:
            List of matching documents with scores
        """
        # For Storage-Optimized endpoints, use VectorSearchClient
        index = self.vs_client.get_index(
            endpoint_name=self.endpoint_name,
            index_name=self.index_name
        )
        
        # Build SQL-like filter string for Storage-Optimized
        filter_str = None
        if filters:
            conditions = []
            for key, value in filters.items():
                if isinstance(value, str):
                    conditions.append(f"{key} = '{value}'")
                elif isinstance(value, list):
                    values_str = ", ".join([f"'{v}'" for v in value])
                    conditions.append(f"{key} IN ({values_str})")
            filter_str = " AND ".join(conditions) if conditions else None
        
        # Default columns if not specified
        if columns is None:
            columns = ["chunk_id", "doc_id", "text", "module", "domain", "doc_name", "source_path"]
        
        # Execute search
        results = index.similarity_search(
            query_text=query_text,
            columns=columns,
            num_results=num_results,
            filters=filter_str,
            query_type=query_type
        )
        
        # Parse results
        documents = []
        if results and hasattr(results, 'get') and results.get('result', {}).get('data_array'):
            data_array = results['result']['data_array']
            
            for row in data_array:
                doc = {}
                for i, col in enumerate(columns):
                    doc[col] = row[i] if i < len(row) else None
                
                # Score is typically the last element
                doc["score"] = row[-1] if len(row) > len(columns) else None
                documents.append(doc)
        
        return documents
    
    def get_index_status(self) -> Dict[str, any]:
        """Get current index status."""
        index_info = self.workspace_client.vector_search_indexes.get_index(
            index_name=self.index_name
        )
        return {
            "name": index_info.name,
            "status": index_info.status,
            "index_type": index_info.index_type,
            "endpoint": index_info.endpoint_name
        }

# Example usage
if __name__ == "__main__":
    print("VectorStore class defined.")
    print("Ready for Vector Search operations!")
````

````python
# src/analysis/llm_orchestrator.py
"""LLM provider abstraction for multi-provider support."""

from abc import ABC, abstractmethod
from typing import Dict, List, Optional
import json
import os

class BaseLLMProvider(ABC):
    """Base class for LLM providers."""
    
    @abstractmethod
    def generate(
        self,
        prompt: str,
        system_prompt: Optional[str] = None,
        max_tokens: int = 4000,
        temperature: float = 0.1,
        response_format: Optional[str] = None
    ) -> str:
        """Generate text completion."""
        pass

class DatabricksLLMProvider(BaseLLMProvider):
    """Databricks Foundation Model API provider."""
    
    def __init__(self, model_name: str = "databricks-dbrx-instruct", endpoint: Optional[str] = None):
        self.model_name = model_name
        self.endpoint = endpoint
        
        # Use Databricks SDK
        from databricks.sdk import WorkspaceClient
        self.client = WorkspaceClient()
    
    def generate(
        self,
        prompt: str,
        system_prompt: Optional[str] = None,
        max_tokens: int = 4000,
        temperature: float = 0.1,
        response_format: Optional[str] = None
    ) -> str:
        """Generate completion using Databricks Foundation Model API."""
        
        # Build messages
        messages = []
        if system_prompt:
            messages.append({"role": "system", "content": system_prompt})
        messages.append({"role": "user", "content": prompt})
        
        # Call API
        response = self.client.serving_endpoints.query(
            name=self.endpoint or self.model_name,
            messages=messages,
            max_tokens=max_tokens,
            temperature=temperature
        )
        
        return response.choices[0].message.content

class OpenAILLMProvider(BaseLLMProvider):
    """OpenAI API provider."""
    
    def __init__(self, model_name: str = "gpt-4", api_key: Optional[str] = None):
        self.model_name = model_name
        self.api_key = api_key or os.getenv("OPENAI_API_KEY")
        
        try:
            import openai
            self.client = openai.OpenAI(api_key=self.api_key)
        except ImportError:
            raise ImportError("Install OpenAI: pip install openai")
    
    def generate(
        self,
        prompt: str,
        system_prompt: Optional[str] = None,
        max_tokens: int = 4000,
        temperature: float = 0.1,
        response_format: Optional[str] = None
    ) -> str:
        messages = []
        if system_prompt:
            messages.append({"role": "system", "content": system_prompt})
        messages.append({"role": "user", "content": prompt})
        
        kwargs = {
            "model": self.model_name,
            "messages": messages,
            "max_tokens": max_tokens,
            "temperature": temperature
        }
        
        if response_format == "json":
            kwargs["response_format"] = {"type": "json_object"}
        
        response = self.client.chat.completions.create(**kwargs)
        return response.choices[0].message.content

class GeminiLLMProvider(BaseLLMProvider):
    """Google Gemini API provider."""
    
    def __init__(self, model_name: str = "gemini-1.5-pro", api_key: Optional[str] = None):
        self.model_name = model_name
        self.api_key = api_key or os.getenv("GEMINI_API_KEY")
        
        try:
            import google.generativeai as genai
            genai.configure(api_key=self.api_key)
            self.model = genai.GenerativeModel(self.model_name)
        except ImportError:
            raise ImportError("Install Gemini: pip install google-generativeai")
    
    def generate(
        self,
        prompt: str,
        system_prompt: Optional[str] = None,
        max_tokens: int = 4000,
        temperature: float = 0.1,
        response_format: Optional[str] = None
    ) -> str:
        # Gemini combines system prompt into user prompt
        full_prompt = prompt
        if system_prompt:
            full_prompt = f"{system_prompt}\n\n{prompt}"
        
        response = self.model.generate_content(
            full_prompt,
            generation_config={
                "max_output_tokens": max_tokens,
                "temperature": temperature
            }
        )
        return response.text

class LLMOrchestrator:
    """High-level LLM orchestration with provider abstraction."""
    
    _providers = {
        "databricks": DatabricksLLMProvider,
        "openai": OpenAILLMProvider,
        "gemini": GeminiLLMProvider,
    }
    
    def __init__(self, provider_name: str = "databricks", **provider_kwargs):
        provider_class = self._providers.get(provider_name.lower())
        if provider_class is None:
            raise ValueError(f"Unknown provider: {provider_name}. Available: {list(self._providers.keys())}")
        
        self.provider = provider_class(**provider_kwargs)
        self.provider_name = provider_name
    
    def generate(
        self,
        prompt: str,
        system_prompt: Optional[str] = None,
        max_tokens: int = 4000,
        temperature: float = 0.1,
        response_format: Optional[str] = None
    ) -> str:
        """Generate completion using configured provider."""
        return self.provider.generate(
            prompt=prompt,
            system_prompt=system_prompt,
            max_tokens=max_tokens,
            temperature=temperature,
            response_format=response_format
        )
    
    def generate_json(
        self,
        prompt: str,
        system_prompt: Optional[str] = None,
        max_tokens: int = 4000,
        temperature: float = 0.1
    ) -> Dict:
        """Generate JSON-formatted response."""
        response_text = self.generate(
            prompt=prompt,
            system_prompt=system_prompt,
            max_tokens=max_tokens,
            temperature=temperature,
            response_format="json"
        )
        
        # Parse JSON
        try:
            return json.loads(response_text)
        except json.JSONDecodeError:
            # Fallback: extract JSON from markdown code blocks
            if "```json" in response_text:
                start = response_text.find("```json") + 7
                end = response_text.find("```", start)
                response_text = response_text[start:end].strip()
            return json.loads(response_text)

# Example usage
if __name__ == "__main__":
    print("LLMOrchestrator class defined.")
    print(f"Supported providers: {list(LLMOrchestrator._providers.keys())}")
````

````python
# src/analysis/impact_analyzer.py
"""Core impact analysis logic using RAG + LLM."""

from typing import Dict, List, Optional
import json

class ImpactAnalyzer:
    """Analyze business requirements and generate impact assessment."""
    
    def __init__(self, vector_store, llm_orchestrator):
        """
        Args:
            vector_store: VectorStore instance for retrieval
            llm_orchestrator: LLMOrchestrator instance for generation
        """
        self.vector_store = vector_store
        self.llm = llm_orchestrator
    
    def analyze_requirement(
        self,
        requirement_text: str,
        domain: Optional[str] = None,
        module: Optional[str] = None,
        top_k: int = 10
    ) -> Dict[str, any]:
        """Perform end-to-end impact analysis.
        
        Args:
            requirement_text: New business requirement description
            domain: Optional domain filter (e.g., "ERP", "Banking")
            module: Optional module filter (e.g., "OrderManagement")
            top_k: Number of relevant documents to retrieve
        
        Returns:
            Structured impact analysis as dict
        """
        
        # Step 1: Retrieve relevant context
        filters = {}
        if domain:
            filters["domain"] = domain
        if module:
            filters["module"] = module
        
        retrieved_docs = self.vector_store.query(
            query_text=requirement_text,
            num_results=top_k,
            filters=filters if filters else None,
            query_type="HYBRID"  # Semantic + keyword
        )
        
        if not retrieved_docs:
            return {
                "error": "No relevant documents found in the repository.",
                "suggestion": "Try broadening your search or checking domain/module filters."
            }
        
        # Step 2: Build context from retrieved docs
        context = self._build_context(retrieved_docs)
        
        # Step 3: Generate structured analysis
        analysis = self._generate_impact_analysis(requirement_text, context)
        
        # Step 4: Add source traceability
        analysis["sources"] = [
            {
                "doc_name": doc.get("doc_name"),
                "module": doc.get("module"),
                "source_path": doc.get("source_path"),
                "relevance_score": doc.get("score")
            }
            for doc in retrieved_docs[:5]  # Top 5 sources
        ]
        
        return analysis
    
    def _build_context(self, documents: List[Dict]) -> str:
        """Build context string from retrieved documents."""
        context_parts = []
        
        for i, doc in enumerate(documents, 1):
            doc_name = doc.get("doc_name", "Unknown")
            module = doc.get("module", "Unknown")
            text = doc.get("text", "")
            score = doc.get("score", 0.0)
            
            context_parts.append(
                f"[Document {i}: {doc_name} | Module: {module} | Relevance: {score:.2f}]\n{text}\n"
            )
        
        return "\n---\n".join(context_parts)
    
    def _generate_impact_analysis(self, requirement: str, context: str) -> Dict[str, any]:
        """Generate structured impact analysis using LLM."""
        
        system_prompt = """You are an expert enterprise change analyst specializing in impact assessment for ERP systems.

Your task is to analyze a new business requirement against existing enterprise documentation and provide a comprehensive impact analysis.

You MUST respond with a valid JSON object following this exact schema:
{
  "requirement_summary": {
    "title": "Brief title",
    "description": "Detailed description",
    "domain": "Domain name",
    "priority": "High/Medium/Low"
  },
  "impacted_modules": [
    {
      "module_name": "Module name",
      "components": ["Component1", "Component2"],
      "impact_level": "High/Medium/Low",
      "rationale": "Why this module is impacted",
      "source_docs": ["doc1.pdf", "doc2.docx"]
    }
  ],
  "validation_changes": [
    {
      "test_type": "Unit/Integration/E2E",
      "test_scope": "What needs testing",
      "rationale": "Why this testing is needed"
    }
  ],
  "regression_scope": ["Area 1", "Area 2"],
  "risks": [
    {
      "risk": "Risk description",
      "severity": "High/Medium/Low",
      "mitigation": "Mitigation strategy"
    }
  ],
  "dependencies": ["Dependency 1", "Dependency 2"],
  "open_questions": ["Question 1", "Question 2"]
}

Be specific, cite relevant information from the provided context, and ensure your analysis is actionable."""
        
        user_prompt = f"""**New Business Requirement:**
{requirement}

**Relevant Enterprise Documentation:**
{context}

**Instructions:**
Analyze the requirement against the provided documentation and generate a comprehensive impact analysis. Follow the JSON schema exactly."""
        
        # Generate JSON response
        try:
            analysis = self.llm.generate_json(
                prompt=user_prompt,
                system_prompt=system_prompt,
                max_tokens=4000,
                temperature=0.1
            )
            return analysis
        except Exception as e:
            return {
                "error": f"Failed to generate analysis: {str(e)}",
                "requirement": requirement
            }
    
    def export_markdown(self, analysis: Dict[str, any]) -> str:
        """Export analysis as markdown report."""
        md = []
        
        # Title
        req_summary = analysis.get("requirement_summary", {})
        md.append(f"# Impact Analysis: {req_summary.get('title', 'Untitled')}")
        md.append("")
        md.append(f"**Domain:** {req_summary.get('domain', 'N/A')}")
        md.append(f"**Priority:** {req_summary.get('priority', 'N/A')}")
        md.append("")
        
        # Description
        md.append("## Requirement Description")
        md.append(req_summary.get('description', 'N/A'))
        md.append("")
        
        # Impacted Modules
        md.append("## Impacted Modules")
        for module in analysis.get("impacted_modules", []):
            md.append(f"### {module.get('module_name')} ({module.get('impact_level')} Impact)")
            md.append(f"**Components:** {', '.join(module.get('components', []))}")
            md.append(f"**Rationale:** {module.get('rationale')}")
            md.append("")
        
        # Risks
        md.append("## Risks & Mitigations")
        for risk in analysis.get("risks", []):
            md.append(f"* **[{risk.get('severity')}]** {risk.get('risk')}")
            md.append(f"  * *Mitigation:* {risk.get('mitigation')}")
        md.append("")
        
        # Dependencies
        md.append("## Dependencies")
        for dep in analysis.get("dependencies", []):
            md.append(f"* {dep}")
        md.append("")
        
        # Open Questions
        md.append("## Open Questions")
        for question in analysis.get("open_questions", []):
            md.append(f"* {question}")
        md.append("")
        
        # Sources
        md.append("## Source Documents")
        for source in analysis.get("sources", []):
            md.append(f"* {source.get('doc_name')} (Module: {source.get('module')}, Relevance: {source.get('relevance_score', 0):.2f})")
        md.append("")
        
        return "\n".join(md)

# Example usage
if __name__ == "__main__":
    print("ImpactAnalyzer class defined.")
    print("Ready for requirement analysis!")
````

# scripts/batch_indexing.py

import sys
sys.path.append('../src')

from pathlib import Path
from ingestion.document_loader import DocumentLoaderFactory
from ingestion.chunking_engine import ChunkingEngine
from storage.vector_store import VectorStore
import uuid
from datetime import datetime

# Configuration
REPOSITORY_PATH = Path("/dbfs/mnt/enterprise-repo")
CATALOG = "main"
SCHEMA = "change_intel"

# Initialize services
vector_store = VectorStore(
    endpoint_name="change-intel-endpoint",
    index_name=f"{CATALOG}.{SCHEMA}.document_index",
    catalog=CATALOG,
    schema=SCHEMA,
    chunks_table="document_chunks"
)

chunker = ChunkingEngine(chunk_size=800, chunk_overlap=100)

# Scan and index
for file_path in REPOSITORY_PATH.rglob("*"):
    if file_path.is_file() and file_path.suffix.lower() in [".pdf", ".docx", ".pptx", ".xlsx", ".txt"]:
        try:
            print(f"Processing: {file_path.name}")
            
            # Load document
            doc_data = DocumentLoaderFactory.load_document(file_path)
            
            # Generate chunks
            chunks = chunker.chunk_text(doc_data["text"])
            
            # Prepare rows for Delta table
            doc_id = str(uuid.uuid4())
            rows = []
            
            for i, chunk in enumerate(chunks):
                rows.append({
                    "chunk_id": f"{doc_id}_chunk_{i}",
                    "doc_id": doc_id,
                    "text": chunk,
                    "doc_name": doc_data["file_name"],
                    "module": "Auto-Detected",  # TODO: Implement auto-detection
                    "domain": "General",
                    "release": "--",
                    "doc_type": doc_data["file_type"],
                    "source_path": str(file_path),
                    "chunk_index": i,
                    "created_at": datetime.now()
                })
            
            # Write to Delta table
            from pyspark.sql import SparkSession
            spark = SparkSession.builder.getOrCreate()
            df = spark.createDataFrame(rows)
            df.write.mode("append").saveAsTable(f"{CATALOG}.{SCHEMA}.document_chunks")
            
            print(f"  ✅ Indexed {len(chunks)} chunks")
            
        except Exception as e:
            print(f"  ❌ Error: {e}")

print("\n🎉 Batch indexing complete!")
print("Run: vector_store.sync_index() to trigger Vector Search sync")
